### Get Embedding

````python
def get_embedding(text:str)->list[float]:

  if not text.strip():
    print("attempted to get embedding for empty text.")
    return []

  embedding=embedding_model.encode(text)
  return embedding.tolist()
````

````python
dataset_df["embedding"] = dataset_df["fullplot"].apply(get_embedding)
````

## Data Retrival



```

 "fields": [{
     "numDimensions": 1024,
     "path": "embedding",
     "similarity": "cosine",
     "type": "vector"
   }]
}
     
```

Use this Atlas search -> Create Atlast Vector Search in MongoDB Vector


````python
def vector_search(user_query,collection):

  query_embedding = get_embedding(user_query)
  print(query_embedding)

  if query_embedding is None:
    return "Invalid query or embeddig is failed"

  pipeline=[

            {
                "$vectorSearch":{

                "index": "vector_index",
                "queryVector": query_embedding,
                "path": "embedding",
                "numCandidates": 150,  # Number of candidate matches to consider
                "limit": 4,  # Return top 4 matches


                }

            },

              {
                 "$project":{

                "fullplot": 1,  # Include the plot field
                "title": 1,  # Include the title field
                "genres": 1,  # Include the genres field
                "score": {"$meta": "vectorSearchScore"},  # Include the search score
                 }

            }

           ]

  result = collection.aggregate(pipeline)
  return list(result)
````

````python
vector_search("what is the best horror movie to watch and why?",collection)
````

````python
def get_search_result(query,collection):

  get_knowledge=vector_search(query,collection)

  search_result=""

  for result in get_knowledge:
        search_result += f"Title: {result.get('title', 'N/A')}, Plot: {result.get('fullplot', 'N/A')}\n"

  return search_result
````

````python
query = "What is the best comedy movie and why?"
````

````python
source_info = get_search_result(query, collection)
````

````python
source_info
````

````python
combined_information = f"Query: {query}\nContinue to answer the query by using the Search Results:\n{source_info}."

print(combined_information)
````

````python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-2b-it")
````

````python
model = AutoModelForCausalLM.from_pretrained("google/gemma-2b-it", device_map="auto")
````

````python
# Moving tensors to GPU
# for preparing the user's query for processing by
# language model and moving it to the GPU for faster computation.
input_ids = tokenizer(combined_information, return_tensors="pt").to("cuda")
````

````python
response = model.generate(**input_ids, max_new_tokens=500)
````

````python
answer = tokenizer.decode(response[0])
````
